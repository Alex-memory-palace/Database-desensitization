```
from pyspark.sql import functions as F

# Select all columns and create a new column 'name' with masked values
masked_df = spark.sql("""
    SELECT 
        *,
        CASE
            WHEN LENGTH(name) > 4 THEN CONCAT(SUBSTRING(name, 1, LENGTH(name) - 4), '****')
            ELSE name
        END AS masked_name
    FROM master_raw.np_engine__c
""")

# Replace the original 'name' column with the new 'masked_name' column
masked_df = masked_df.withColumn("NAME", F.col("masked_name")).drop("masked_name")

# Display the updated DataFrame
masked_df.display()
```
